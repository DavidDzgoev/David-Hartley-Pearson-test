```python
import pandas as pd
import time
```

```python
start_time = time.time()
def main():
    def dc_split(line):
        A = line.split(',')
        date = A[2]
        year = date[0:4]
        cost = float(A[7])
        dcv = [year,cost]
        return dcv

    def cpy(co,data1):
        df = []
        data = []
        co = str(co)
        with open(f"{co}.csv") as f:
            for line in f:
                a = dc_split(line)
                data.append(a)
        for year in range(2015,2020):
            closes = []
            close_changes = []
            for i in range(len(data)):
                if str(year) == data[i][0]:
                    close = data[i][1]
                    closes.append(close)
            for i in range(len(closes)-1):
                close_change = closes[i+1] - closes[i]
                close_changes.append(close_change)
            change_min = min(close_changes)
            df.append(change_min)
        data1.append(df)
    data1=[]
    companies = [
        "ALRS",
        "GAZP",
        "CHMF",
        "ADR.FIVE",
        "GMKN",
        "LKOH",
        "MGNT",
        "MTSS",
        "NLMK",
        "NVTK",
        "ROSN",
        "SBER",
        "SNGS",
        "TATN",
        "YNDX"
    ]
    for co in companies:
        cpy(co,data1)
    df = pd.DataFrame(data=data1, index = companies, columns = range(2015,2020))
    return df

main()
print("--- %s seconds ---" % (time.time() - start_time))
main()
```
