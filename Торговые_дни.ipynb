```python
import pandas as pd
import time
```

```python
start_time = time.time()
def main():
    def year_split(line):
        A = line.split(',')
        date = A[2]
        year = date[0:4]
        return year

    def dpy(co,day):
        data = []
        co = str(co)
        with open(f"{co}.csv") as f:
            for line in f:
                a = year_split(line)
                data.append(a)     
        day1 = []
        for year in range(2015,2020):
            td = data.count(str(year))
            day1.append(td)
        day.append(day1)
    day=[]
    companies = [
        "ALRS",
        "GAZP",
        "CHMF",
        "ADR.FIVE",
        "GMKN",
        "LKOH",
        "MGNT",
        "MTSS",
        "NLMK",
        "NVTK",
        "ROSN",
        "SBER",
        "SNGS",
        "TATN",
        "YNDX"
    ]
    for co in companies:
        dpy(co,day)
    df = pd.DataFrame(data=day, index = companies, columns = range(2015,2020))
    return df
main()
print("--- %s seconds ---" % (time.time() - start_time))
main()
```
