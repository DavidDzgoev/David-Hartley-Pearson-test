```python
import matplotlib.pyplot as plt
import numpy as np
import datetime
import time
```

```python
start_time = time.time()
def main():
    def date_split(line):
        A = line.split(',')
        dat = A[2]
        year = int(dat[0:4])
        mon = int(dat[4:6])
        day = int(dat[6:])
        date = datetime.date(year,mon,day)
        return date

    def close_split(line):
        A = line.split(',')
        close = float(A[7])
        return close

    def gca(co):
        dates = []
        close = []
        co = str(co)
        with open(f"{co}.csv") as f:
            for line in f:
                a = date_split(line)
                dates.append(a)
                b = close_split(line)
                close.append(b)
        plt.xlabel("Временной период")
        plt.ylabel("Цена акции")
        plt.plot(dates,close,color="green",linewidth=0.5,alpha = 0.5)
    gca('MGNT')
main()
print("--- %s seconds ---" % (time.time() - start_time))
main()
```
