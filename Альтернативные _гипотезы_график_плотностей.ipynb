```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import random
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
n = stats.norm(0,1)
s = stats.logistic(0,2/3)
t = stats.t(4)

x = np.linspace(-5,5)
y1 = [n.pdf(i) for i in x]

y2 = [s.pdf(i) for i in x]

y3 = [t.pdf(i) for i in x]

fig, ax = plt.subplots()

plt.grid()

ax.plot(x, y1, label = 'Norm(0,1)')
ax.plot(x, y2, label = 'Logistic(0,2/3)')
ax.plot(x, y3, label = 'T(4)')

ax.legend()

plt.show()
```
